```python
import csv  
import spacy  
import pandas as pd  

# Load the English language model  
nlp = spacy.load("en_core_web_sm")  

# Path to the CSV file  
file_path = 'spam.csv'  

# Read the CSV file with proper encoding  
df = pd.read_csv(file_path, encoding='latin1')  


# Function to lemmatize text and remove stop words  
def lemaStopWord(text_series):  
    filtered_words = []  

    
    for text in text_series: 

        doc = nlp(text)  # Process the text with spaCy  

        # Filter out stop words and lemmatize the remaining tokens  
        filtered = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]

        concatenated_lemmas = ' '.join(filtered)  


        filtered_words.append(concatenated_lemmas)  
    
    return filtered_words  







# Apply the function to the DataFrame text series  
filtered_results = lemaStopWord(df["v2"]) 

X =  filtered_results
# Encode labels
y = df["v1"].map({'ham': 0, 'spam': 1})


print(y)




```

```python
from sklearn.model_selection import train_test_split


# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y)

print(X_test)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer


# CountVectorizer()
vectorizer = TfidfVectorizer()  



X_train_vectorized = vectorizer.fit_transform(X_train).toarray()
X_test_vectorized = vectorizer.transform(X_test).toarray()

 
```

```python
from sklearn.neural_network import MLPClassifier  

# Initialize and train the MLPClassifier  
clf = MLPClassifier(random_state=1, max_iter=300)  
clf.fit(X_train_vectorized, y_train) 
```

```python
from sklearn.metrics import accuracy_score  
from sklearn.metrics import classification_report  

# Make predictions on the test set  
y_pred = clf.predict(X_test_vectorized) 

# Calculate the accuracy  
accuracy = accuracy_score(y_test, y_pred) * 100 


print(f"Model accuracy on test set: {accuracy:.2f}")  

cr = classification_report(y_test, y_pred) 

print (cr)
```
